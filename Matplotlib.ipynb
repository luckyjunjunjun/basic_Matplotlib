# 简单的函数设置

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 1
# 简单的函数展示，记得在末尾加上plt.show()
x = np.linspace(-3, 3, 50)
y1 = x**2
y2 = 2 * x + 1

plt.figure()
plt.plot(x, y1)

plt.show()
```

# 格式设置

```python
import matplotlib.pyplot as plt
import numpy as np

# 没有Figure 2
# （Figure3）
plt.figure(num = 3, figsize = (8, 5))           # num修改的是figure的序号

x = np.linspace(-3, 3, 50)
y1 = x**2
y2 = 2 * x + 1

# 设置线条格式
plt.plot(x, y2)
plt.plot(x, y1, color = 'red', linewidth = 10.0, linestyle = '--')

# 修改坐标轴的范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# 修改坐标轴的名字
plt.xlabel('I am x')
plt.ylabel('I am y')

# 换坐标点：plt.xticks(坐标点的设置)   坐标点的设置 = np.linspace(始位置, 终位置, 分隔位置数)
new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)

# 用名称代替数字
plt.yticks([-2, -1.8, -1, 1.22, 3],
            ['really bad', 'bad', 'normal', 'good', 'really good'])



# 设置坐标轴的位置与格式
# gca = 'get current axis'，把轴拿出来
ax = plt.gca()
# spines 脊梁
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom') # 把x轴上的坐标点放在底部
ax.yaxis.set_ticks_position('left') # 把y轴上的坐标点放在左边
ax.spines['bottom'].set_position(('data', 0)) # 把x轴放在在y轴的0处
ax.spines['left'].set_position(('data', 0))  # 把y轴放在x轴的0处
# 还有其他定位方式：outward，axes等等



# Legend 图例
l1, = plt.plot(x, y2, label = 'up') # label为名字
l2, = plt.plot(x, y1, color = 'red', linewidth = 1.0, linestyle = '--', label = 'down')
plt.legend() # 打印出图例
# 用于打印出个性化的图例, handles用于传入，label用于命名，loc用于选定位置（best时会自动找一个位置）
# 定义l1, l2时，需要在等号和名称之间加逗号，才能正常传入
# 可以只打印一条线的图例
plt.legend(handles = [l1, l2], labels = ['aaa', 'bbb'], loc='best')



#Annotation 注解
x0 = 1
y0 = 2 * x0 + 1
#plt.scatter(x, y1) # 布局散点在线上
plt.scatter(x0, y0, s = 50, color = 'b') # 只将(x0, y0)变为散点,s为尺寸，b为蓝色
plt.plot([x0, x0], [y0, 0], 'k--', lw = 2.5) # k代表黑色，--为虚线，lw即line width线条宽度

# method 1
# xycoords使得x,y坐标基于data，即x0, y0这个点，
# xytext代表开始打印注解的位置
# textcoords选定注解的点
# fontsize代表注解的大小（包括线）
# arrowstyle选定注解线样式
# connectionstyle决定弧度和角度
plt.annotate(r'$ 2x + 1 = %s $' %y0, xy = (x0, y0), xycoords = 'data',      # 可以用$等符号修改字体（特殊格式） 
             xytext = (+1.2, 1), #textcoords = 'offset points',
             fontsize = 16, arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3, rad = 0.2'))  # 以data为基准

# method 2
plt.text(-1, 1, r'$These\ are\ some\ text,\ \mu\ \sigma_i\ \alpha_t$',
         fontdict = {'size': 12, 'color': 'r'})



# tick 能见度 —— 运用举例：坐标点被粗线条挡住的时候
# bbox即label后面的框框
# alpha修改的是颜色透明度
# 以x,y轴上的坐标点举例
for label in ax.get_xticklabels() + ax.get_yticklabels():  # 把x，y轴上的点拿出来（选中）
    label.set_fontsize(12)      # 修改尺寸
    label.set_bbox(dict(facecolor = 'blue', edgecolor = 'None', alpha = 0.1))  # 修改坐标点所在小方框的背景颜色，


plt.show()
```

# 散点图

```python
import matplotlib.pyplot as plt
import numpy as np

# 散点图（ 位于Figure 4 ）
plt.figure()

n = 1024
X = np.random.normal(0, 1, n) # x和y用数字代替(随机生成)，平均值是0，方差是1，生成n = 1024个数
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)  # 修改颜色, T代表颜色数量值，不同的数量值代表不同颜色

plt.scatter(X, Y, s = 75, c = T, alpha = 0.5) # 看一张漂亮的图，c为color
plt.scatter(np.arange(5), np.arange(5)) # 看一条线的散点

#plt.xlim((-1.5, 1.5))
#plt.ylim((-1.5, 1.5))

# 隐藏所有的ticks，只留下图片
plt.xticks(())
plt.yticks(())

plt.show()
```

# 柱状图

```python
import matplotlib.pyplot as plt
import numpy as np

# 柱状图 (位于 Figure5 )
plt.figure()
n = 12                  # 12个柱状图
X = np.arange(n)        # 生成 0 ~ 11 的数字
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) # 根据X产生随机变量，从0.5 ~ 1.0， 每个n都产生一个
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

# bar即柱状图,正负号区分柱体向上和向下，颜色默认时为蓝色，需要修改时可以输入对应颜色的代码
plt.bar(X, +Y1, facecolor = '#9999ff', edgecolor = 'white')
plt.bar(X, -Y2, facecolor = '#ff9999', edgecolor = 'white')

# 添加文字
# zip用于将X, Y1传到x, y, 不加zip时每一步只会输出一个值, zip后才会输出x和y两个值
for x, y in zip(X, Y1):
    # 前两个参数用于设定位置，第三个参数为柱状图的值, ha与va分别是横向、纵向对齐方式
    plt.text(x, y + 0.05, '%.2f' %y, ha = 'center', va = 'bottom')

for x, y in zip(X, Y2):
    # 前两个参数用于设定位置，第三个参数为柱状图的值, ha与va分别是横向、纵向对齐方式
    plt.text(x, -y - 0.05, '-%.2f' %y, ha = 'center', va = 'top')

plt.ylim((-1.2, 1.2))


plt.show()
```

# 等高线图

```python
import matplotlib.pyplot as plt
import numpy as np

# Contours 等高线图 ( 位于 Figure6 )
plt.figure()

# 拿x, y两个点，计算它们的高度
def f(x, y):
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 - y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
# 先定义一个网格（底图）
X, Y = np.meshgrid(x, y)

# 前三个参数输入xyz三个坐标轴
# 8 这个参数代表分成10部分 （最少为0，分2部分）
# cmap(cm) —— color map —— 拿给定的数字，调用成对应数字的颜色, hot —— 暖色调
plt.contourf(X, Y, f(X, Y), 8, alpha = 0.75, cmap = plt.cm.hot)   # 设定颜色

# 描等高线
# 还可以用 linewidth 等参数修改格式
C = plt.contour(X, Y, f(X, Y), 8, colors = 'black')

#数字描述
# 传入， 是否写在线上(false时线会直接穿过数字)，尺寸
plt.clabel(C, inline = True, fontsize = 10)

plt.show()
```

# 输出色彩与数字的对应图片

```python
import matplotlib.pyplot as plt
import numpy as np

# 打印图像 ( Figure7 )

# 用数字作为伪图片
plt.figure()
a = np.array([0.313, 0.365, 0.423,
              0.365, 0.439, 0.525,
              0.423, 0.525, 0.651]).reshape(3, 3)

# 首个参数用于传入
# lower 与 upper 颜色放置方向相反
# interpolation 用于选择图像样式，种类详见以下链接
# http://matplotlib.org/examples/images_contours_and_fields/interpolation_methods.html
plt.imshow(a, interpolation = 'nearest', cmap = 'bone', origin = 'upper')

# colorbar() 输出色彩与数字的对应图例, shrink用于压缩colorbar的长度
plt.colorbar(shrink = 0.9)

plt.show()
```

# 3D图

```python
import matplotlib.pyplot as plt
import numpy as np

# 3D图 ( Figure8 )
from mpl_toolkits.mplot3d import Axes3D     # 添加3D坐标轴显示

fig = plt.figure()      # ( Figure8 )
ax = Axes3D(fig, auto_add_to_figure = False)        # 添加3D坐标轴
fig.add_axes(ax)        # 3.9以上版本需加上此句才会出图

X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)         # 用于生成网格采样点矩阵

R = np.sqrt(X ** 2 + Y ** 2)     # R用于计算X和Y，得到Z的高度值
Z = np.sin(R)

# 正式开始画图
# 第四、五个参数代表线与线之间的跨度，越大网就越稀疏
ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = plt.get_cmap('rainbow'))

# zdir用于设定投影的方向, offset设定具体投到多远
ax.contourf(X, Y, Z, zdir = 'z', offset = -2, cmap = 'rainbow')
ax.set_zlim(-2, 2)

plt.show()
```

# 分格显示的3种方式

## method 1

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 10

# Subplot 分格显示
# 使用 method2 时需加上如下import
import matplotlib.gridspec as gridspec


# method1: subplot2grid
plt.figure()
# 3行3列，从第0行第0列开始, colspan(rowspan)为跨度, 默认时跨度为1
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan = 3, rowspan = 1)
ax1.plot([1, 2], [1, 2])
ax1.set_title('ax1_title')     # 用此格式代替原本的plt.title()等

ax2 = plt.subplot2grid((3, 3), (1, 0), colspan = 2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan = 2)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))

plt.show()
```

## method 2

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 11

# method2: gridspec
plt.figure()
# 与 ':' 相关的参数，初项序号为1; 不与 ':' 相关的参数, 初项序号为0
gs = gridspec.GridSpec(3, 3)         # 3行3列
ax1 = plt.subplot(gs[0, :])         # ':' 表示占了所有列（整一行）
ax2 = plt.subplot(gs[1, :2])        # ':2' 表示从初始列占到第2列
ax3 = plt.subplot(gs[1:, 2])         # '1:' 表示从初始行后（即第2行）开始占，占到最后一行
ax4 = plt.subplot(gs[-1, 0])
ax5 = plt.subplot(gs[-1, -2])


plt.show()
```

## method 3

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 12

# method3: easy to define structure
# plt.figure()     不需要
# share 时坐标轴会被各个子图共用，不share则每个子图都有自己的轴
# f 是图的代号，也可以用其他名称代替
f, ((ax11, ax12), (ax21, ax22)) = plt.subplots(2, 2, sharex = True, sharey = True)
ax11.scatter([1, 2, 3], [1, 2, 3])      # 散点图


plt.show()
```

# 图中图

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 13

# 图中图
fig = plt.figure()
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 4, 2, 5, 8, 6]

left, bottom, width, height = 0.1, 0.1, 0.8, 0.8        # 边框位置（在figure中所占比例）
ax1 = fig.add_axes([left, bottom, width, height])      # 传入参数
ax1.plot(x, y, 'r')     # r即red
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

left, bottom, width, height = 0.2, 0.6, 0.25, 0.25       # 修改位置，放在大figure中
ax2 = fig.add_axes([left, bottom, width, height])      # 传入参数
ax2.plot(y, x, 'b')     # b即blue, 将x和y反过来举例
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title inside 1')

# 这种方法不需要像以上那样set，会紧跟着axes
plt.axes([0.6, 0.2, 0.25, 0.25])
plt.plot(y[::-1], x, 'g')                  # y[::-1]相当于y[-1: -len(a)-1: -1],从最后一个元素到第一个元素复制一遍; 'g'即green
plt.xlabel('x')
plt.ylabel('y')
plt.title('title inside 2')

plt.show()
```

# 次坐标轴

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 14

# 次坐标轴
x = np.arange(0, 10, 0.1)
y1 = 0.05 * x**2
y2 = -1 * y1

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()              # 用镜面把ax1的数据反过来
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')

ax1.set_xlabel('X data')
ax1.set_ylabel('Y1', color = 'g')
ax2.set_ylabel('Y2', color = 'b')


plt.show()
```

# 动画

```python
import matplotlib.pyplot as plt
import numpy as np

# Figure 15

# Animation 动画
from matplotlib import animation

fig, ax = plt.subplots()

x = np.arange(0, 2 * np.pi, 0.01)
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x + i / 10))             # 更新y的数据
    return line,

def init():
    line.set_ydata(np.sin(x))             # 设置初始帧
    return line,

# frames即插入的时间点数（帧数）, init_func为初始帧, interval即更新的频率（几ms更新一次）
# blit是否更新整张图的店（否的话，只更新变化的点；是的话，没变化的点也会更新）
ani = animation.FuncAnimation(fig = fig, func = animate, frames = 100, init_func = init, interval = 20, blit = True)


plt.show()
```
